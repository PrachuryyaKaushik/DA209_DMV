**Dimensions in Arrays** <br>

```python
# 0-dimensional array (only 1 item)
a = np.array(42)
# 1-Dimensional array
b = np.array([1, 2, 3, 4, 5])
# 2-Dimensional array
c = np.array([[1, 2, 3], [4, 5, 6]])
# 3-Dimensional array
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)

print('Zero Dimension:\n',a)
print('One Dimension:\n',b)
print('Two Dimension:\n',c)
print('Three Dimension:\n',d)
```

## **Matplotlib**
Matplotlib was developed by John D. Hunter in 2003, it enables users to graphically represent data, facilitating easier analysis and understanding.

```python
# Let's install and import the library first
!pip install matplotlib
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y)
plt.show()
```

**Lets give a little bit of more information**

```python
plt.plot(x,y)
plt.xlabel("X-Values")
plt.ylabel("Y-axis")
plt.title("Basic Graph")
plt.show()
```

**We can plot more than one in the same plot**

```python
z = np.arange(1,3.1,0.1)
#print(z)
plt.plot(x,y)
plt.plot(z,z**2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Graph")
plt.show()
```

**But what does this plots mean?**

```python
z = np.arange(1,3.1,0.1)
plt.plot(x,y,label='2x')
plt.plot(z,z**2,label='z^2')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Graph")
plt.legend()
plt.show()
```

## **Seaborn**
Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures.

```python
# Let's install and import the library first
!pip install seaborn
import seaborn as sns
```

```python
# Set seaborn theme
sns.set_theme(style="whitegrid")

# Define data
x = np.arange(1, 3.1, 0.1)
y = 2 * x
z = x  # using the same range as x
z_squared = z ** 2

# Plot the data
plt.figure(figsize=(8, 5))
sns.lineplot(x=x, y=y, label="2x", marker="o", linewidth=2)
sns.lineplot(x=z, y=z_squared, label="z²", marker="s", linewidth=2)

# Add labels and title
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.title("Basic Graph using Seaborn", fontsize=14)
plt.legend(title="Functions")
plt.tight_layout()
plt.show()
```

## **Plotnine**
Plotnine is a Python implementation of the Grammar of Graphics — a powerful and flexible system for creating statistical graphics based on layered principles originally developed in ggplot2 for R.

```python
# Let's install the library first
!pip install plotnine
```

```python
import numpy as np
import pandas as pd
from IPython.display import display

from plotnine import (
    ggplot, aes, geom_line, geom_point,
    labs, theme_minimal, theme, element_text, scale_color_manual
)

# Prepare data
x = np.arange(1, 3.1, 0.1)
df = pd.DataFrame({
    "x": list(x) * 2,
    "y": list(2 * x) + list(x ** 2),
    "Function": ["2x"] * len(x) + ["z²"] * len(x)
})

# Build the plot using grammar of graphics
plot = (
    ggplot(df, aes(x="x", y="y", color="Function", shape="Function")) +  # Aesthetics
    geom_line(size=1.5) +                                                # Geometry: lines
    geom_point(size=2.5) +                                               # Geometry: points
    scale_color_manual(values={"2x": "#1f77b4", "z²": "#ff7f0e"}) +      # Custom colors
    labs(
        title="Graph of Functions 2x and z²",
        x="X-axis",
        y="Y-axis",
        color="Function",
        shape="Function"
    ) +                                                                  # Labels
    theme_minimal() +                                                   # Theme
    theme(
        axis_title=element_text(size=12),
        plot_title=element_text(size=14, weight='bold', ha='center'),
        legend_title=element_text(size=10),
        legend_text=element_text(size=9)
    )
)

print(plot)

# At the end of your code
display(plot.draw())
```
